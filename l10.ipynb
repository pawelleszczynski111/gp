```python
def count_mean(numbers_list):
  sum=0
  for el in numbers_list:
    sum+= el
  mean= sum/ len(numbers_list)
  return mean
```

```python
test_list=[2,2,3,4,5,5,6]
mean=count_mean(test_list)
print(mean)
```
