```python
def count_mean(numbers_list):
  sum=0
  for el in numbers_list:
    sum+= el
  mean= sum/ len(numbers_list)
  return mean
```

```python
def count_mode(numbers_list):
  elements=[]
  most_often=[]

  for el in numbers_list:
    counter=0
    for el2 in numbers_list:
      if el == el2:
        counter+=1
    elements.append(el)
    most_often.append(counter)

  modes=[]
  for i in range(len(elements)):
    if most_often[i] == max(most_often):
      modes.append(elements[i])

  modes = list(set(modes))
  return modes



```

```python
pajaki=[48, 49, 51, 52]
ender_dragony=[0, 0, 100, 100]
szkielety= [10, 30, 70, 90]
zombie=[49, 50, 50, 51]
```

```python
def count_variance(group):
  mean=statistics.mean(group)
  variance= statistics.variance(group)
  return variance, mean
```

```python
groups={
    'Pająki': pajaki,
    'Ender Dragony': ender_dragony,
    'Szkielety': szkielety,
    'Zombie': zombie
}
```

```python
variances={}
means={}

for name, group in groups.items():
  var, mean= count_variance(group)
  variances[name]= var
  means[name]= mean
  print(f'{name} - Średnia: {mean}, Wariancja: {var}')
```
