```python
from transformers import AutoTokenizer, AutoModelForCausalLM, pipeline
```

```python
tokenizer= AutoTokenizer.from_pretrained('flax-community/papuGaPT2')
model= AutoModelForCausalLM.from_pretrained('flax-community/papuGaPT2')
generator= pipeline('text-generation', model=model, tokenizer= tokenizer)
```

```python
text= generator('Sztuczna inteligencja to', max_length=20,temperature=0.6)

print(text[0]['generated_text'])
```

```python
"""
max_length lub max_new_tokens -- długość wygenerowanego tekstu

temperature -- kreatywnośc / losowosć -> wyższa -- bardziej kreatywny model

do_sample -- True - losowe / False - deterministyczne

top_k -- wybiera tokeny z listy najlepszych tokenów

top_p -- wybiera tokeny które razem dają p pewności

repetition_penalty -- kara za powtarzanie tokenów

num_beams -- liczba ścieżek

num_return_sequences -- liczba wygenerowanych wersji tekstu

length_penalty -- kara za dłuższe odpowiedzi

bad_words_ids -- lista tokenów, które model powinien omijać

pad_token_id -- Token wypełnienia

eos_token_id -- Token końca tekstu



 """
```
